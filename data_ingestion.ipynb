# Ingest data from Pic Database

### import libraries

```python
import numpy as np
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

import os
import random

import cv2
from imutils import paths


import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from keras.preprocessing.image import img_to_array
```

### set target dimensions of the image

```python
PIC_DIMS = (96, 96, 3)
```

```python
# initialize datasets
data = []
labels = []
```

### extract file paths and shuffle (with a constant seed)

```python
path = os.getcwd() # script directory
path = os.path.join(path,'PicDb') #Pictures directory
print(path)
```

```python

PicPaths = sorted(list(paths.list_images(path)))
random.seed(42)
random.shuffle(PicPaths)
```

```python
print("Total number of snake images (venomous + nonvenemous) :", len(PicPaths))
```

### Extractions

```python
counter =0
for picpath in PicPaths:
    print(picpath)
    pic = cv2.imread(picpath)
    print(picpath, pic.shape)
    
    pic = cv2.resize(pic, (PIC_DIMS[1],PIC_DIMS[0]))
    pic = img_to_array(pic)
    print(pic.shape)
    data.append(pic)
    
    counter= counter+1
    label = picpath.split(os.path.sep)[-2]
   # print(label)
    labels.append(label)
```

```python
print("total images: for training", counter )
```

```python
count_nonvenom = 0
count_venom = 0

for count in range(len(labels)):
    if labels[count]=='picdb-nonvenom':
        count_nonvenom = count_nonvenom+1
    else:
        count_venom = count_venom+1

print("total venomous snake images: ",count_venom)
print("total nonvenomous snake images: ",count_nonvenom)
```

```python
lb = LabelBinarizer()
labels_t = lb.fit_transform(labels)

```

```python
print(labels[100],labels_t[100])
```

```python
data_norm = np.array(data, dtype="float") / 255.0

data_norm.shape
```

```python
print("extracted imageset datasizes") 
print("\nimageset size in MB:" , data_norm.nbytes/(1024*1024))
print("\nlabelset size in MB:" , labels_t.nbytes/(1024*1024))

```
